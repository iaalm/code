#等灰度颜色

根据RGB转灰度的公式`Gray=R*0.299+G*0.587+B* 0.114`，生成具有相同灰度的各个颜色，并描绘在同一张图上，以此来直观的表示灰度与RGB的关系。
下图中之所以用RB来作为坐标轴，是因为G的系数最大，能够较为平缓的展开颜色，看起来更加方便。


```python
%pylab inline
```

```python
# 看一下imread的结果
imread('/Users/simon/Projects/Lena.png')
```

```python
# Gray=R*0.299+G*0.587+B* 0.114

c = 4  #行数列数
for i in range(c * c):
    Grey = i / c / c * 256
    pic = zeros((256,256,3),dtype=uint8)
    for R in range(256):
        for B in range(256):
            G = (Grey-R*0.299-B*0.114)/0.587
            if G >= 0 and G < 256:
                pic[R][B][0] = R
                pic[R][B][1] = G
                pic[R][B][2] = B
    subplot(c,c,i+1)
    imshow(pic)
```

```python
c = 4  #行数列数
for i in range(c * c):
    Grey = i / c / c * 64 + 128 - 16
    pic = zeros((256,256,3),dtype=uint8)
    for R in range(256):
        for B in range(256):
            G = (Grey-R*0.299-B*0.114)/0.587
            if G >= 0 and G < 256:
                pic[R][B][0] = R
                pic[R][B][1] = G
                pic[R][B][2] = B
    subplot(c,c,i+1)
    imshow(pic)
```
